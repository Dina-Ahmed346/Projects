# using the iris dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
```

```python

iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)


iris['species'] = iris_data.target

iris.head(40)
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris.isnull().sum()
```

```python
iris.drop_duplicates(inplace=True)
```

```python
X = iris
```

```python
iris.info()
```

```python
col = X.columns
col
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
X = ms.fit_transform(X)
```

```python
X
```

```python
X = pd.DataFrame(X, columns=[col])
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
print(labels[:200])
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X, kmeans.labels_)
print("The average silhouette score is :", silhouette_avg)
```

```python

from matplotlib import pyplot as plt


plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=kmeans.labels_)

#Display the plot.
plt.show()
```

```python
#Using grid search
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_clusters': [2, 3, 4, 5, 6],
    'init': ['k-means++'],
    'max_iter': [100, 200, 300],
    'n_init': [10, 20]
}

# Create the KMeans model
kmeans = KMeans()

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=kmeans, param_grid=param_grid, cv=5, verbose=1)

# Fit the model to the data
grid_search.fit(X)

# Print the best parameters and score
print("Best parameters:", grid_search.best_params_)

# Get the best model
best_kmeans = grid_search.best_estimator_

# Use the best model to make predictions
predictions = best_kmeans.predict(X)

```

```python
silhouette_avg = silhouette_score(X, predictions)
print("Silhouette score:", silhouette_avg)
```

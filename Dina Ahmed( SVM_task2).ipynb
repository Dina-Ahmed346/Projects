```python
categorical_columns = rt.select_dtypes(include=['object']).columns
numerical_columns = rt.select_dtypes(include=['int64', 'float64']).columns
```

```python
categorical_columns
```

```python
for column in categorical_columns:
    rt[column].fillna(rt[column].mode()[0], inplace=True)
```

```python
## filling the numerical columns with the median
for column in numerical_columns:
    rt[column].fillna(rt[column].median(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X = rt.drop(['target'], axis='columns')
y = rt.target
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=121)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
